**Line Graph**

```r

speed<-c(10,20,30,45,100)
distance<-c(100,200,300,450,1000)
part<-data.frame(speed,distance)
print(part)

```

```r
plot(part,main="Car's speed and distance")
plot(part,main="Car's speed and distance",type="o")
plot(part,main="Car's speed and distance",type="p")
plot(part,main="Car's speed and distance",type="h")
plot(part,main="Car's speed and distance",type="l")
plot(part,main="Car's speed and distance",type="s")
plot(part,main="Car's speed and distance",type="s",xlab="speed",ylab="distance",col='red')

#multiple lines in line graph
d<-c(13,62,67,8798,345,,789,0,900,45,121,567,232,34345,133,434,21,23,434,45,4677,76,23,345,676,567,7)
plot(part,main="car dataset",type="l",xlab="speed",ylab="distance",col="green")
lines(d,type="o",col="blue")
```

**Scatter Plot**

```r
#scatter plot
speed<-c(10,20,30,45,100)
distance<-c(100,490,300,345,200)
part<-data.frame(speed,distance)
print(part)
plot(x=part$speed,y=part$distance,main="Car's speed and distance")
plot(x=part$speed,y=part$distance,main="Car's speed and distance",sub="Data")
plot(x=part$speed,y=part$distance,main="Car's speed and distance",sub="Data",col="red")
plot(x=part$speed,y=part$distance,main="Car's speed and distance",sub="Data",col="green",frame.plot=FALSE)
plot(x=part$speed,y=part$distance,main="Car's speed and distance",sub="Data",col="red",xlab="speed",ylab="distance",axes=FALSE)
```

**Box Plot**

```r
#box plot
speed<-c(10,20,30,45,100)
distance<-c(100,490,300,345,200)
part<-data.frame(speed,distance)
part
boxplot(part)
boxplot(part,main="Speeed and distance",xlab="speed", ylab="distance",col=c("green","red"))


p<-c(10,20,30,45,100)
boxplot(p,main="employee ages",col="red")
boxplot(p,main="employee ages",col="red",notch=TRUE)
boxplot(p,main="employee ages",col="red",varwidth=TRUE)
boxplot(p,main="employee ages",col="red",varwidth=TRUE,border="green")
boxplot(p,main="employee ages",col="red",varwidth=TRUE,border="green",horizontal=TRUE)

```

**Bar Chart**

```r
#Bar Chart
part<-c(10,20,30,45,100)
nam<-c("Shyam","Rohan","Mohan","Sadashiv","Rehan")
#case 1
barplot(part,main="Employees",names.arg=nam)

#case 2
barplot(part,main="Employees",col=c("red","blue","green","yellow","violet"))

#case 3
barplot(part,main="Employees",col=c("red","blue","green","yellow","violet"))

#case 4
barplot(part,main="Employees",col=c("red","blue","green","yellow","violet"),names.arg=nam,xlab="names of employees",ylab="age of employees")

#case 5
barplot(part,main="Employees",density=10,col=c("red","blue","green","yellow","violet"),names.arg=nam,xlab="names of employees",ylab="age of employees")
barplot(part,main="Employees",density=10,angle=180,col=c("red","blue","green","yellow","violet"),names.arg=nam,xlab="names of employees",ylab="age of employees")

#case 6
barplot(part,main="Employees",density=10,col=c("red","blue","green","yellow","violet"),names.arg=nam,xlab="names of employees",ylab="age of employees",horiz=TRUE)
legend("topright",fill=c("red","blue","green","yellow","violet"),legend=c("Shyam","Rohan","Mohan","Sadashiv","Rehan"),bg="white")


#case 7
#Stacked bar plot
v1<-c(1,2,3,4,5,6,7,8,9)
v2<-matrix(v1,nrow=3,ncol=3)
barplot(v2,col=c("red","blue","green"), main="Stacked bar chart")

#case 8
#group bar plot side by side
barplot(values,col=c("red","blue","green"), main="Stacked bar chart",beside=TRUE)


```

**Pie Chart**

```r
#pichart
a<-c(12,56,87,62,10)
l<-c("danceing","singing","acting","Playing","cooking")

#case 1
pie(a,labels=l)

#case 2
pie(a,labels=l,edge=10,main="Hobbies")

#case 3
pie(a,labels=l,radius = -1 ,main="Hobbies")

pie(a,labels=l,radius = 1 ,main="Hobbies")

#case 4
pie(a,labels=l,edge=10,radius = -1 ,main="Hobbies")

#case 5
pie(a,labels=l,clockwise=TRUE ,main="Hobbies")
pie(a,labels=l,clockwise=FALSE ,main="Hobbies")

#case 6
pie(a,labels=l,density=20 ,main="Hobbies")
pie(a,labels=l,density=100 ,main="Hobbies")

#case 7
pie(a,labels=l,density=20, angle=90,main="Hobbies")
pie(a,labels=l,density=20, angle=45,main="Hobbies")
pie(a,labels=l,density=20, angle=135,main="Hobbies")

#case 8
pie(a,labels=l,col=c("red","blue","green","yellow","violet"),main="Hobbies")
pie(a,labels=l,col=c("red","blue","green","yellow","violet"),border="red",main="Hobbies")
pie(a,labels=l,border="red",main="Hobbies")

#case 9
#legend(x, y, legend, fill, col, bg, lty, cex, title, text.font, bg)\
pie(a,labels=l,border="red",main="Hobbies")
legend("topright",fill=c("red","blue","green","yellow","violet"),legend=c("danceing","singing","acting","Playing","cooking"),bg="white")

#case 10
#install.packages("plotrix")
library(plotrix)
data <- c(19, 21, 54, 12, 36, 12)
names<- c("Happy","sad","angry","dousy","joyish")
pie3D(data,labels=names,col=c("pink","yellow","sky blue","orange","black"),main="Emotions and their frequencies")
legend("topright",fill=c("red","blue","green","yellow","violet"),legend=c("pink","yellow","sky blue","orange","black"))


```
